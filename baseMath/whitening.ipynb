```python
import matplotlib.pyplot as plt
import numpy as np
xsrc = np.loadtxt('pcaData.txt') 
print(xsrc.shape) # 45 sample, 2 dimension
print(np.mean(xsrc, axis=1))
print(np.std(xsrc, axis=1))
plt.xlim(-4,4); plt.ylim(-4,4)
```

```python
x = xsrc.T
plt.scatter(x[:,0], x[:,1], marker='o', color='', edgecolors='b')
plt.show()
```

```python
# standarlize
x -= np.mean(x, axis=0)
x /= np.std(x, axis=0)
plt.xlim(-4,4); plt.ylim(-4,4)
plt.scatter(x[:,0], x[:,1], marker='o', color='', edgecolors='b')
plt.show()
```

```python
# PCA 
x = xsrc.T
x -= np.mean(x, axis=0)
convar = np.dot(x.T, x) / x.shape[0]
u,s,v = np.linalg.svd(convar)
xrot = np.dot(x, u.T)
plt.xlim(-4,4); plt.ylim(-4,4)
plt.scatter(xrot[:,0], xrot[:,1], marker='o', color='', edgecolors='b')
plt.show()
```

```python
# PCA white
e = 1e-5
xPCAwhite = xrot * (np.diag(1./np.sqrt(np.diag(s)+e)))
plt.xlim(-4,4); plt.ylim(-4,4)
plt.scatter(xPCAwhite[:,0], xPCAwhite[:,1], marker='o', color='', edgecolors='b')
plt.show()
```

```python
# ZCA white
xZCAwhite = xPCAwhite.dot(u)
plt.xlim(-4,4); plt.ylim(-4,4)
plt.scatter(xZCAwhite[:,0], xZCAwhite[:,1], marker='o', color='', edgecolors='b')
plt.show()
```
