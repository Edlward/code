```python
# 把时间转换为以分钟为单位
def returnmin(time):
    clickT = time % 10000
    clickT = clickT.values  / 100 * 60 + clickT.values % 100
    return clickT
clickTiMin = returnmin(train['clickTime'])
converTiMin = returnmin(train['conversionTime'])
# type(clickTiMin)
dfTime = pd.DataFrame({'clickTiMin':clickTiMin, 'converTiMin':converTiMin})
dfTime = dfTime.fillna(0)
# print type(dfTime)
# dfTime.head()
dfTime['plusTime'] = dfTime['converTiMin'] - dfTime['clickTiMin']
# dfTime[dfTime['plusTime'] > 0]['plusTime'] = 0
```

```python
v = dfTime[dfTime['plusTime'] > 0]
# 激活的conversionTime与clickTime的时间差
sns.distplot(v['plusTime'],kde=False)
```

```python
# 点击时间分布
sns.distplot(dfTime['clickTiMin'],kde = False)
fig = plt.figure()
dfTime['label'] = train['label']
means = dfTime.groupby('clickTiMin')['label'].mean()
plt.scatter(x=means.index, y=means.values)
# conversionTime只有label是1时才会出现，所以并没有需要绘图
```
